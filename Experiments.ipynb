```python
import matplotlib.pyplot as plt

def plot_predictions(means, stds, i=0, e=1, y_lim=[0, 0.4], title="", include_forth=False):
  x = []
  for t in range(means.shape[0]):
    x += [10 ** (t + 1)]
  plt.xscale('log')
  plt.ylim(y_lim)

  plt.xticks(fontsize=16)
  plt.yticks(fontsize=16)

  if include_forth:
    color = 'orange'
    mean = means[:,3,i]
    std = stds[:,3,i]
    plt.plot(x, mean, label=f"Line {color}", color=color, linewidth=3)  # Mean line
    plt.fill_between(x, mean - std, mean + std, alpha=0.2, color=color)  # Shaded area

    mean = means[:,3,e]
    std = stds[:,3,e]
    plt.plot(x, mean, label=f"Line {color}", color=color, linestyle='--', linewidth=3)  # Mean line
    plt.fill_between(x, mean - std, mean + std, alpha=0.2, color=color)  # Shaded area

  color = 'green'
  mean = means[:,0,i]
  std = stds[:,0,i]
  plt.plot(x, mean, label=f"Line {color}", color=color, linewidth=3)  # Mean line
  plt.fill_between(x, mean - std, mean + std, alpha=0.2, color=color)  # Shaded area

  mean = means[:,0,e]
  std = stds[:,0,e]
  plt.plot(x, mean, label=f"Line {color}", color=color, linestyle='--', linewidth=3)  # Mean line
  plt.fill_between(x, mean - std, mean + std, alpha=0.2, color=color)  # Shaded area

  color = 'red'
  mean = means[:,1,i]
  std = stds[:,1,i]
  plt.plot(x, mean, label=f"Line {color}", color=color, linewidth=3)  # Mean line
  plt.fill_between(x, mean - std, mean + std, alpha=0.2, color=color)  # Shaded area

  mean = means[:,1,e]
  std = stds[:,1,e]
  plt.plot(x, mean, label=f"Line {color}", color=color, linestyle='--', linewidth=3)  # Mean line
  plt.fill_between(x, mean - std, mean + std, alpha=0.2, color=color)  # Shaded area

  color = 'blue'
  mean = means[:,2,i]
  std = stds[:,2,i]
  plt.plot(x, mean, label=f"Line {color}", color=color, linewidth=3)  # Mean line
  plt.fill_between(x, mean - std, mean + std, alpha=0.2, color=color)  # Shaded area

  mean = means[:,2,e]
  std = stds[:,2,e]
  plt.plot(x, mean, label=f"Line {color}", color=color, linestyle='--', linewidth=3)  # Mean line
  plt.fill_between(x, mean - std, mean + std, alpha=0.2, color=color)  # Shaded area

  plt.title(title, fontsize=20)
  plt.xlabel("Graph size", fontsize=20)
  plt.ylabel("Class probability", fontsize=20)
```
